```python
import networkx as nx
```

```python
G = nx.read_edgelist('network.tsv', nodetype=int)
```

```python
evaluation = {}
invalid = 0

with open('candidates.txt') as file:
    working_list = []
    for line in file:
        y = [int(x) for x in line.split()]
        if G.has_edge(y[0], y[1]) == True:
            print((y[0], y[1]), 'Invalid')
            invalid += 1
            continue
        working_list.append((y[0], y[1]))
    
    print(invalid)
    results = nx.jaccard_coefficient(G, working_list)
    for a,b,c in results:
        evaluation[(a,b)] = c
```

```python
evaluation
```

```python
final = sorted(evaluation.items(), key = lambda x: x[1], reverse=True)[:50000]
```

```python
final
```

```python
with open('part1.txt', 'w') as file:
    for x in final:
        file.write('{}\t{}\n'.format(x[0][0], x[0][1]))
```
