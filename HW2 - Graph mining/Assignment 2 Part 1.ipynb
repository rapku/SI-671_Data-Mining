```python
seedlist = [50,104,36,400,1239,8321,232,53,1223,6]
# degree_ref = 1
```

```python
results = []

for seed in seedlist:
    print('Running seed {}'.format(seed))
    random.seed(seed)
    reservoir = []
    G = nx.Graph()

    with open('network.tsv') as file:
        i = 0
        for line in file:
            i+=1
            y = line.split()
            if len(reservoir) < no_edges:
                reservoir.append((int(y[0]), int(y[1])))
            else:
                reserv = int(random.random()*i)
                if reserv < no_edges:
                    reservoir[reserv] = (int(y[0]), int(y[1]))
    print('Reservoir sampling complete')
    
    G = nx.Graph()
    for x in reservoir:
        G.add_edge(x[0], x[1])
    del reservoir
    print('Graph complete')
        
#     high_deg = [x[0] for x in G.degree() if x[1]>degree_ref]
    high_deg = list(G.nodes())
    print(len(high_deg))
    
    iter = 0
    for x in high_deg:
        similarity(x, G)
        if iter%10000 == 0:
            results = sorted(results, key = lambda x: x[1], reverse=True)[:150000]
            print('Cleaning of results done, iteration {}'.format(iter))
        iter += 1
    results = sorted(results, key = lambda x: x[1], reverse=True)[:150000]
    print('Results completed')
    
fileres = open('candidates.txt', 'w')
for x in results:
    fileres.write('{}\t{}\n'.format(x[0][0], x[0][1]))
fileres.close()
    
```
